### This notebook handles anonymization of excels and files

```python
import os
import numpy as np
import pandas as pd
wd = r'N:\AUG-ImageRepository\anonymized_xml\OCT_depth_project\data'
os.listdir(wd)
```

```python
#get mapping xlsx
patient_mapping_path = os.path.join(wd, "patient_ID.xlsx") 
#get labels files
labels_diagnoses_prozeduren_path = os.path.join(wd, 'labels_diagnoses_prozeduren.xlsx') 
labels_VISUS_path = os.path.join(wd, 'labels_VISUS.xlsx') 
labels_IOL_path = os.path.join(wd, 'labels_IOL.xlsx') 
#load csv's into pandas
patient_mapping_pd = pd.read_excel(patient_mapping_path)
labels_diagnoses_prozeduren_pd = pd.read_excel(labels_diagnoses_prozeduren_path)
labels_VISUS_pd = pd.read_excel(labels_VISUS_path)
labels_IOL_pd = pd.read_excel(labels_IOL_path)
```

```python
#merging the old with new ID's
anonym_labels_diagnoses_prozeduren_pd = patient_mapping_pd.merge(labels_diagnoses_prozeduren_pd\
                         , left_on = 'P_ID_DB', right_on = "P_ID_DB", how = "left").drop(columns = "P_ID_DB")
anonym_labels_VISUS_pd = patient_mapping_pd.merge(labels_VISUS_pd\
                         , left_on = 'P_ID_DB', right_on = "P_ID_DB", how = "left").drop(columns = "P_ID_DB")
anonym_labels_IOL_pd = patient_mapping_pd.merge(labels_IOL_pd\
                         , left_on = 'P_ID_DB', right_on = "P_ID_DB", how = "left").drop(columns = "P_ID_DB")
```

```python
#saving names
labels_DIAG_PROZ_save_path = os.path.join(wd, 'anom_labels_DIAG_PROZ.xlsx') 
labels_VISUS_save_path = os.path.join(wd, 'anom_labels_VISUS.xlsx') 
labels_IOL_save_path = os.path.join(wd, 'anom_labels_IOL.xlsx') 

# Create a Pandas Excel writer using XlsxWriter as the engine.
writer_DIAG_PROZ = pd.ExcelWriter(labels_DIAG_PROZ_save_path, engine='xlsxwriter')
writer_VISUS = pd.ExcelWriter(labels_VISUS_save_path, engine='xlsxwriter')
writer_IOL = pd.ExcelWriter(labels_IOL_save_path, engine='xlsxwriter')

#save pd's
anonym_labels_diagnoses_prozeduren_pd.to_excel(writer_DIAG_PROZ)
anonym_labels_VISUS_pd.to_excel(writer_VISUS)
anonym_labels_IOL_pd.to_excel(writer_IOL)

#close writers
writer_DIAG_PROZ.save()
writer_VISUS.save()
writer_IOL.save()
```
